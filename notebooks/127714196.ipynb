```python
%load_ext autoreload
%autoreload 2

# External imports
import matplotlib.pyplot as plt
import polars as pl
import seaborn as sns

# Local imports
from investment_analysis.constants import colors as co
from investment_analysis.constants import feature_names as fn
from investment_analysis.features.parse_timestamp_to_datetime import parse_timestamp_to_datetime
```

```python
df = pl.read_csv('../data/processed/127714196.csv')
```

```python
df = parse_timestamp_to_datetime(df)
```

```python
from datetime import datetime

market_crashes = [
    (datetime(2020,2,20),datetime(2020,3,23)), # ⁠COVID-19 Crash
    (datetime(2022,1,3),datetime(2022,10,12)), # ⁠Bear Market
    (datetime(2023,3,8),datetime(2023,3,13)),  # Regional Banking Crisis
]
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(24, 13.5))

ax = sns.lineplot(df, x=fn.DATETIME, y=fn.CLOSE, color=co.HONOLULU_BLUE)

for start, end in market_crashes:
    ax.axvspan(start, end, color=co.IMPERIAL_RED_MAIN, alpha=0.1)

plt.show()
```
